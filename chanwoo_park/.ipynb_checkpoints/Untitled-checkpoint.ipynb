```python
from collections import deque
```

```python
dq = deque()
for e in kill_data:
#     print(e._D)
    try:
        dq.append([
             e._D,
            {"x":e.assistant.location.x,"y": e.assistant.location.y, "z":e.assistant.location.z},
             e.assistant.name,
             e.common.is_game,
             e.damage_causer_name,
             e.damage_reason,
             {"x":e.killer.location.x,"y": e.killer.location.y, "z":e.killer.location.z},
             e.killer.name,
             {"x":e.victim.location.x,"y": e.victim.location.y, "z":e.victim.location.z},
             e.victim.name,
        ])
    except:
        try:
            dq.append([
                e._D,
                {
                    "x": -1,
                    "y": -1,
                    "z": -1,
                },
                'None',
                e.common.is_game,
                e.damage_causer_name,
                e.damage_reason,
                {"x":e.killer.location.x,"y": e.killer.location.y, "z":e.killer.location.z},
                e.killer.name,
                {"x":e.victim.location.x,"y": e.victim.location.y, "z":e.victim.location.z},
                e.victim.name,
            ])
        except:
            dq.append([
                e._D,
                {
                    "x": -1,
                    "y": -1,
                    "z": -1,
                },
                'None',
                e.common.is_game,
                e.damage_causer_name,
                e.damage_reason,
                {
                    "x": -1,
                    "y": -1,
                    "z": -1,
                },
                'None',
                {"x":e.victim.location.x,"y": e.victim.location.y, "z":e.victim.location.z},
                e.victim.name,
            ])

        
df = pd.DataFrame(dq, columns=['time', 'asist_loc','asist_name', 'phase', 'dmg_cause', 'dmg_reason', 'kill_loc', 'kill_name', 'vict_loc', 'vict_name'])
```
