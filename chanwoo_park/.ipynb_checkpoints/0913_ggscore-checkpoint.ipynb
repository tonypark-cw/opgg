```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url_base = 'https://ggscore.com'
```

```python
url_id = 'https://ggscore.com/en/pubg/player?region=1&country=1&t='
```

```python
pls = {6: {'2heart', 'ayoh', 'indigo', 'kayle', 'menteul', 'minsung', 'pio', 'seongjang', 'silky', 'under'},
    2: {'alphaca', 'ej', 'hwan2da', 'inonix', 'loki', 'orange', 'sparrow', 'yunjong', 'zenith'}, 
        3: {'adder', 'dg98', 'dowon', 'kali', 'maczora', 'piky', 'rex'}, 
            5: {'akan', 'aloha', 'hellen', 'k1nch'}, 
                0: {'dino', 'gujamyoung', 'smoke'}, 
                    4: {'alone', 'alyu', 'apxb', 'donghyun', 'geonhu', 'laura', 'minuda', 'spacebin', 'yun', 'zamil'}, 
                        1: {'aka', 'chanahri', 'h0ziboy', 'kasa', 'mxxu', 'nojam', 'poodle', 'tanggle'},}
```

```python
stats_list = []
# pls : players
# pl : player label
for pl in pls:
    for p in pls.get(pl):
        res = req.get(url_id+p)  # 먼저 선수 이름을 조회
        html = bs(res.text, 'html.parser')
        p_id = html.select('table.mtable tr.t-item a')
        try:
            nick_name = p_id[0]['href']
        except:
            continue
        
        res = req.get(url_base+nick_name)  # 먼저 선수 nick_name을 받아와서 검색시작
        html = bs(res.text, 'html.parser')

        stats = html.select('div.table-responsive table tr td span.counter')

        p_data = html.select('div.pInf table td:nth-child(2)')
        p_data_info = ['NAME', 'AGE', 'COUNTRY','TEAM','EARNED','MATCHES']

        t_list=[]
        for data in p_data:
            t_list.append((data.text).replace('\n','').replace('\t',''))
    #     print(t_list)
        stat_dict = {}
        for i in range(len(p_data_info)):
#             print(p_data_info[i])
            try:
                if p_data_info[i] == 'MATCHES':
#                     print("들어옴")
                    min_matches = t_list[i].split('/')[0]
#                     print(min_matches)
                    min_matches = min_matches.strip()
                    min_matches = min_matches.replace('-', '0')
#                     print(min_matches)
                    stat_dict['MATCHES'] = min_matches
                else:
                    stat_dict[p_data_info[i]] = t_list[i]
            except:
                pass
        s_list = ['FORM', 'RATING', 'MATCH PERFORMANCE', 'PRECISION' , 'SURVIVAL', 'DISTANCE', 'KILLS', 'EXPERIENCE']

    #         
        stat_dict['NICKNAME'] = p
        for i, stat in enumerate(stats):
            stat_dict[s_list[i]] = (stat.text)

        stats_list.append(stat_dict)
```
