```python
tm_target = os.listdir('tournaments/')
tm_target.remove('.DS_Store')
tm_target
```

```python
df_origin = pd.DataFrame()
for tournament_id in tm_target:
    battles = os.listdir('tournaments/'+tournament_id+"/")
    for battle in battles:
        if '_' in battle and battle.endswith('csv'):
            df = pd.read_csv('tournaments/'+tournament_id+"/"+battle)
            df['tournament_id'] = tournament_id
            df['match_id'] = battle.split('_')[1].split('.')[0]
#             print(df)
            df_origin = pd.concat([df_origin, df])
#         break
#     break
```
