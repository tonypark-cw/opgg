```python
def save_image(player_name):
    url = 'https://liquipedia.net'

    res = req.get(url+'/pubg/'+player_name)
    html = bs(res.content,'html.parser')
    
    try:
        image_url = html.select_one('span.flag a')['title']
    except:
        img_dict[player_name] = ''
    else:
        img_dict[player_name] = (url + image_url)
    print(player_name, img_dict[player_name])
    time.sleep(10)
#         with open('선수사진/'+player_name+'.jpg', 'wb') as f:
#             f.write(res_img.content)
```

```python
df = pd.read_csv('2020_tournament_table_수정후 - 2020_tournament_table_수정후.csv')
```

```python
df
```

```python
for player_name in list(df['player_name']):
    save_image(player_name)
```

```python
for k,v in img_dict.items():
    if not v:
        save_image(k)
```

```python

df_img = pd.DataFrame({'player_name':img_dict.keys(),'imagme_url':img_dict.values()})    
```

```python
df_result = pd.merge(df, df_img)
```
