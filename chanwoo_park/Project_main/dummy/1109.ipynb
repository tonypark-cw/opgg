```python
player_stats_csvs = [tele for tele in os.listdir('./dummy/tournaments_telemetry/') if 'player' in tele]
```

```python
player_stats_csvs
```

```python
df_stack = pd.DataFrame()
for player_stats_csv in player_stats_csvs:
    df = pd.read_csv("./dummy/tournaments_telemetry/"+player_stats_csv)
    df['tournament_id'] = player_stats_csv.split('_')[0]
    df_stack = df_stack.append(df)
```

```python
df_stack['player_name'] = df_stack['player'].str.split('_').str[1]
```

```python
url = 'https://liquipedia.net'
player_name = 'Pio'
res = req.get(url+'/pubg/'+player_name)
html = bs(res.content,'html.parser')
country_code = html.select_one('div span.flag + a')['title'].split(":")[1]
country_code
```

```python
img_dict = {}
cc_dict = {}
```

```python
def save_image(player_name):
    url = 'https://liquipedia.net'

    res = req.get(url+'/pubg/'+str(player_name))
    html = bs(res.content,'html.parser')
    
    try:
        image_url = html.select_one('a.image img')['src']
    except:
        img_dict[player_name] = ''
    else:
        img_dict[player_name] = (url + image_url)
    
    try:
        country_code = html.select_one('div span.flag + a')['title'].split(":")[1]
    except:
        cc_dict[player_name] = ''
    else:
        cc_dict[player_name] = country_code
    print(player_name,cc_dict[player_name], img_dict[player_name])
    time.sleep(60)
#         with open('선수사진/'+player_name+'.jpg', 'wb') as f:
#             f.write(res_img.content)
```

```python
for player_name in df_stack['player_name'].unique():
    if player_name not in list(img_dict.keys()):
        save_image(player_name)
```

```python

df_img = pd.DataFrame({'player_name':img_dict.keys(),'imagme_url':img_dict.values(), 'country_code':cc_dict.values()})    
```

```python
df_img
```

```python
df_result = pd.merge(df_stack, df_img)
```

```python
df_result.to_csv('2020_선수_요약_url포함.csv')
```
