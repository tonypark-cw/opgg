1. K-Means로 하면 영향을 준것을 같이 볼수 있다.(10개 안쪽으로) -> 너무 많으면 해석이 어렵다. 클러스터링도 해석이 굉장히 중요하니까.
- 해석을 해보면서 한번 확인을 하면 훨씬 좋을것 같다.

2. 매 경기에서 잘한 선수를 뽑아준다.(MVP) , 지금 데이터 셋으로 0~10 점까지 점수를 매긴다면, 어떻게 부여하게 될까를 고민해보쟈...
- 리니어 모형으로 확인을 해도 괜찬고
- 감에 의존해서 해도 괜찬다.
- 중요하게 사용된 피처들을 통해 구분을 해봐도 괜찬다.
- y값을 생존시간, 킬수 혹은 순위, 거리, 데미지, 이런것들로 할때엔 어떨까.
- 승리한 팀인데, 팀원이 먼저 탈락한경우 이 팀원은 노이즈가 될 수 있다.

파생변수로 생존시간 대비 데미지 값을 찾아보자.

스코어 모델을 모델링으로 만드는 경우도 있는데 모델링으로 넣으면 의도치않은 값이 들어가기도 한다.
- 누군가에게 보여졌을때 체감이 되는게 가장 중요하다. -> 설득력이 중요하다.
- 스코어 수식을 만들어서 선수들을 적용했는데, 선수들 점수 분포가 어느한쪽에 치우쳐져있다면, 좋은 모델은 아닐것이다.
