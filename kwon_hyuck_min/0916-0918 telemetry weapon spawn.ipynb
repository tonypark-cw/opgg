```python
def weapon_spawn_list(tour_id):
    api_key = "eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJqdGkiOiI4N2RhZGJiMC1lMmJlLTAxMzktNzg3Yi0yN2U2ZTUzOGUwMTUiLCJpc3MiOiJnYW1lbG9ja2VyIiwiaWF0IjoxNjI5MzM4MTYyLCJwdWIiOiJibHVlaG9sZSIsInRpdGxlIjoicHViZyIsImFwcCI6ImJhdHRsZWdyb3VuZF9tIn0.2GJwJPXu0OjvsaSdc6MeoXy9AgKngxQlL102pdRt1SI"
    pubg = PUBG(api_key=api_key, shard="tournament")
    match_ids = get_match_id(tour_id)
    x_e = []
    y_e = []
    item_name_e = []
    x_m = []
    y_m = []
    item_name_m = []
    
    global erangel_weapon_df
    global miramar_weapon_df
    
    for n in match_ids:
        match = pubg.match(n)
        telemetry = match.get_telemetry('telemetry_json')
        pick_events = telemetry.filter_by('log_item_pickup')
        if telemetry.map_id() == 'Baltic_Main': 
            for p in range(len(pick_events)):
                if pick_events[p]['item']['item_id'] == 'Item_Weapon_SCAR-L_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] == 'Item_Weapon_SKS_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_BerylM762_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_AK47_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] == 'Item_Weapon_Mini14_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] == 'Item_Weapon_Kar98k_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_FNFal_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_HK416_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_M24_C':
                    item_name_e.append(pick_events[p]['item']['item_id'])
                    x_e.append(pick_events[p]['character']['location']['x'])
                    y_e.append(pick_events[p]['character']['location']['y'])
                    
            erangel_weapon_df = pd.DataFrame(list(zip(item_name_e, x_e , y_e)), columns=['weapon', 'x', 'y'])
            
        elif telemetry.map_id() == 'Desert_Main':
            for p in range(len(pick_events)):
                if pick_events[p]['item']['item_id'] == 'Item_Weapon_SCAR-L_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] == 'Item_Weapon_SKS_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_BerylM762_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_AK47_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] == 'Item_Weapon_Mini14_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] == 'Item_Weapon_Kar98k_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_FNFal_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_HK416_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                elif pick_events[p]['item']['item_id'] =='Item_Weapon_M24_C':
                    item_name_m.append(pick_events[p]['item']['item_id'])
                    x_m.append(pick_events[p]['character']['location']['x'])
                    y_m.append(pick_events[p]['character']['location']['y'])
                    
            miramar_weapon_df = pd.DataFrame(list(zip(item_name_m, x_m , y_m)), columns=['weapon', 'x', 'y'])
            
            
    return erangel_weapon_df, miramar_weapon_df
```
